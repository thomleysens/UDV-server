# Automated process

We are going to use our own Python methods and classes to get accessibility data and "*holes map*" from OpenStreetMap features and accessibility measures. 

We use a *master* Python file ```get_holes.py``` with parameters in a JSON schema file ```park_params.json```. 


> ***WARNING ON OSM PARK DATA: there is an "*access*" tag that could be "*private*" but it seems most of data have no value for this tag or is not correct.***

## Example on Lyon Metropole

### Get graph if needed

We must use a bounding box larger than the bounding box used for the query on OSM (*via Overpass*) to get parks in order to measure accessibility from parks located on the edges of the bounding box's query. 

> ***Use ```graph_utils``` methods only if you don't already have the required nodes and edges JSON files. So no need to run the cell below if you already have these files*** 

````python
import osmnx as ox

from graph_utils import graph_to_df, graph_with_time

# Parameters
# Get graph from Lyon Metropole bounding box
bbox = [4.55, 45.38, 5.2, 45.97] #from Geofabrik Tile Calculator, [left, bottom, right, top]

# Get graph from OSM data
graph = ox.graph_from_bbox(
    bbox[3],
    bbox[1],
    bbox[2],
    bbox[0],
    network_type="walk"
) #bbox => north, south, east, west

# Add time as weight in graph (based on walk speed)
walk_distance = 5000 #in meters for 1 hour trip
graph = graph_with_time(graph, walk_distance)

#Write graph to disk as 2 files (nodes and edges)
nodes_path = "./data/Lyon_metropole/park_case/nodes.json"
edges_path = "./data/Lyon_metropole/park_case/edges.json"
graph_to_df(graph, edges_path, nodes_path)
````

### Parameters

> ***Warning: we use key/value pair leisure/park to get OSM data. Data could be incomplete and must be checked.***

Here is what's in the ```params.json``` file:

```json
{
"osm_bbox_query" : [45.53, 4.62, 45.97, 5.11],
"osm_key_query" : "leisure",
"osm_value_query" : "park",
"graph_nodes_jsonfile" : "./data/Lyon_metropole/park_case/nodes.json",
"graph_edges_jsonfile" : "./data/Lyon_metropole/park_case/edges.json",
"epsg_graph" : 4326,
"epsg_input" : 4326,
"epsg_metric" : 2154,
"output_features_layername" : "parks",
"output_isolines_layername" : "isolines",
"output_buffered_isolines_layername" : "buffered_isolines",
"output_buffered_isolines_union_layername" : "buffered_isolines_union",
"output_folder" : "./data/Lyon_metropole/park_case",
"output_format" : "geopackage",
"trip_times" : [5,10,15],
"distance_buffer" : 40,
"lat" : "x",
"lon" : "y"
}

```

Let's see each parameter:


| Parameter | Description | Type | Example |
|:---------|:-----------|:----:|:-------:|
| ```osm_bbox_query``` | Bounding box for OSM query, must be (SOUTH, WEST, NORTH, EAST) and EPSG 4326 | **array** | *[45.53, 4.62, 45.97, 5.11]* |
| ```osm_key_query``` | Key string for OSM query | **str** | *"leisure"* |
| ```osm_value_query``` | Value string for OSM query | **str** | *"park"* |
| ```graph_nodes_jsonfile``` | Graph nodes JSON file | **str** | *"my_directory/graph/nodes.json"* |
| ```graph_edges_jsonfile``` | Graph edges JSON file | **str** | *"my_directory/graph/edges.json"* |
| ```epsg_graph``` | Spatial reference system (SRS) or coordinate reference system (CRS) of graph | **int** | *4326* |
| ```epsg_input``` | SRS/CRS of (OSM) query | **int** | *4326* |
| ```epsg_metric``` | SRS/CRS of (OSM) query | **int** | *2154* |
| ```output_features_layername``` | Layer name for OSM requested features | **str** | *"parks"* |
| ```output_isolines_layername``` | Layer name for isolines | **str** | *"isolines"* |
| ```output_buffered_isolines_layername``` | Layer name for buffered isolines | **str** | *"buffered_isolines"* |
| ```output_buffered_isolines_union_layername``` | Layer name for union of buffered isolines | **str** | *"buffered_isolines_union"* |
| ```output_folder``` | Path to output folder | **str** | *"my_directory/output"* | 
| ```output_format``` | Format for output ("*shapefile*", "*geojson*", "*geopackage*", "*all*") | **str** | *"all"* | 
| ```trip_times``` | Array of durations in minutes | **array** | [5,10,15] |
| ```distance_buffer``` | Distance in meters for buffering isolines to get polygons | **int** | 40 |
| ```lat``` | Name of the latitude values in graph files (*default: "x"*) | **str** | *"x"* |
| ```lon``` | Name of the longitude values in graph files (*default: "y"*) | **str** | *"y"* |


We use a validator to check types.

### Output explanations

We obtain Shapefiles, GeoJSON files or a [GeoPackage (*an open format for geospatial information*)](http://www.geopackage.org/) file that you can, for example, open with [QGIS](https://www.qgis.org/en/site/), but also with [GeoPandas](http://geopandas.org/), among others (*see the last section of this notebook for examples*). 

Files or layers(*GeoPackage format*):
* ***isolines***: walkable accessibility from points (*around parks in our examples*). The file contains Viridis colors values for each duration and each duration. 
* ***buffered isolines***: same as previous but with buffering to get polygons instead of polylines
* ***union of buffered isolines***: same as previous but with union of all polygons
* ***OSM polygons***: polygons layer from OSM query (*here parks*)
* ***Layer with union of OSM polygons and buffered isolines for each duration***, for example:
    * 5 minutes (*called ```5``` in our examples*)
    * 10 minutes (*called ```10``` in our examples*)
    * 15 minutes (*called ```15``` in our examples*)
